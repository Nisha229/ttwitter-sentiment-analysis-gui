```python
import time
import tweepy
from flair.models import TextClassifier
from flair.data import Sentence
from google.colab import userdata
from tweepy.errors import TooManyRequests

# Load Bearer Token
bearer_token = userdata.get("BEARER_TOKEN")
client = tweepy.Client(bearer_token=bearer_token)

# Load Flair sentiment model
classifier = TextClassifier.load("sentiment")

def preprocess_text(text):
    return text.replace('\n', ' ').strip()

def get_sentiment(text):
    sentence = Sentence(text)
    classifier.predict(sentence)
    return sentence.labels[0].value, sentence.labels[0].score

# Real-time tweet sentiment loop
while True:
    try:
        print("🔄 Fetching new tweets...")
        response = client.search_recent_tweets(query="#crypto", max_results=10)

        if response.data:
            for tweet in response.data:
                original_tweet = tweet.text
                clean_tweet = preprocess_text(original_tweet)
                sentiment, score = get_sentiment(clean_tweet)

                print('-------------------------Tweet----------------------')
                print(original_tweet)
                print('----------------------------------------------------')
                print('Sentiment:', sentiment, f"(Score: {score:.2f})\n")
        else:
            print("No new tweets found.")

        # Wait 60 seconds before the next batch
        time.sleep(60)

    except TooManyRequests:
        print("⚠️ Rate limit hit. Waiting 15 minutes...")
        time.sleep( 6*60)  # Wait 15 minutes
```

```python
# Suppose this is the tweet you want to test:
tweet_text = "Crypto is booming today! 🚀💰"

# Clean it
clean = preprocess_text(tweet_text)
print("Cleaned Tweet:", clean)

# Predict sentiment
sentiment, score = get_sentiment(clean)
print("Sentiment:", sentiment, f"({score:+.3f})")


```

```python
### gradio code
# --- INSTALL DEPENDENCIES ---
!pip install gradio transformers tweepy

# --- IMPORT LIBRARIES ---
import gradio as gr
import tweepy
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import numpy as np
from collections import Counter

# --- LOAD EMOTION DETECTION MODEL ---
model_name = "j-hartmann/emotion-english-distilroberta-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
emotion_labels = ['anger', 'disgust', 'fear', 'joy', 'neutral', 'sadness', 'surprise']

# --- LOAD TWITTER API KEYS ---
import google.colab.userdata as userdata
bearer_token = userdata.get('BEARER_TOKEN')  # Set using Colab > Tools > Secrets
client = tweepy.Client(bearer_token=bearer_token)

# --- EMOTION PREDICTION FUNCTION ---
def get_emotion(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True)
    with torch.no_grad():
        logits = model(**inputs).logits
    probs = torch.nn.functional.softmax(logits, dim=1).numpy()[0]
    top_idx = np.argmax(probs)
    return emotion_labels[top_idx]

# --- MAIN FUNCTION TO FETCH AND ANALYZE TWEETS ---
def analyze_tweets(hashtag):
    result = []
    emotion_counts = Counter()

    try:
        response = client.search_recent_tweets(query=hashtag, max_results=10)
        if response.data:
            for tweet in response.data:
                emotion = get_emotion(tweet.text)
                emotion_counts[emotion] += 1
                result.append(f"🧠 {tweet.text}\n🎭 Emotion: {emotion.upper()}\n")

            stats = "\n".join([f"{k.capitalize()}: {v}" for k, v in emotion_counts.items()])
            return "\n".join(result), stats
        else:
            return "No tweets found.", ""

    except Exception as e:
        return f"Error: {e}", ""
```
